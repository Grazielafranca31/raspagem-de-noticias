Iniciando pelo site da **Fiquem Sabendo**

```python
#Fiz a requisição e printei pra saber se estava funcionando sem erros
linkfs = 'https://fiquemsabendo.com.br/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
requisicaofs = requests.get(linkfs, headers=headers)
print(requisicaofs.status_code)
```

```python
print(requisicaofs.content.decode())
```

```python
#Utilizei o BeautifulSoup para fazer a raspagem da parte que eu queria do site, que eram as notíciais
html_page = requisicaofs.text
soupfs = BeautifulSoup(html_page, 'html.parser')
```

```python
noticias = soupfs.findAll('h3',{'class':'gdlr-blog-title'})
```

```python
#li só a variável para ver se estava tudo certo
noticias
```

```python
#fiz um laço para agregar todas as manchetes e links em uma lista
noticias_fs=[]
for noticia in noticias:
  manchete=noticia.find('a').text
  link=noticia.find('a').get('href')
  noticias_fs.append([manchete,link])

```

```python
#fiz um DataFrame para separar em colunas a manchete e link
dffs=pd.DataFrame(noticias_fs, columns=['Manchete','Link'])
```

```python
#utilizei .assign para acrescentar uma coluna ao meu DataFrame com o nome do site
dffs = dffs.assign(site = 'Fiquem Sabendo')
```

```python
#Fiz outros .assign para utilizar o datetime e acrescentar outra coluna
dffs = dffs.assign(data_extracao=date.today()) 
```

```python
#Lendo o Dataframe final dessa página pra saber se ficou tudo certo
dffs
```

```python
#Achei esse outro código na internet e fiz os laços por ele 
words = dict()
arquivo = '/content/newstatu.csv'

with open(arquivo, 'r', encoding='utf-8') as manchetes:  # aqui eu coloquei a variável arquivo em str
    reader = csv.reader(manchetes)
    for ler in reader:

        for palavra in ler[1].split():

            # Se existir é verdadeiro
            if palavra in words:
                words[palavra] = int(words[palavra]) + 1
            else:
                words[palavra] = 1

# Imprime o dicionario
print(words)

# For do dicionario por linha
for chave, valor in words.items():
    print('Palavra= ' + str(chave)+((20-len(str(chave)))*' ') + ' - Quantidade= ' + str(valor))
```

```python
#Precisei remover as stopwords
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("portuguese"))
```

```python
#esse dicionário está recebendo os filtrados
filtered={}
for chave, valor in words.items():
    if chave.lower() not in stop_words and len(chave) > 1:
         filtered.update({chave.lower() : valor})
         print(chave, valor)
print(filtered) #palavras sem stopwords

```

```python
#ordenando por decrescente
for words in sorted(filtered, key = filtered.get, reverse=True):
  print(words, filtered[words])

```

```python
#Agora coloquei todas as palavras que apareciam 3 vezes ou mais e printei em uma frase 
for a, b in sorted(filtered.items(), reverse=True):
  if b>=3:
    print(f'A palavra {a} aparece {b} vezes em manchetes')
```
